## Face Detection

```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()

    # Load Haar cascade
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Check if the cascade is loaded properly
    if face_cascade.empty():
        raise IOError("Cannot load Haar cascade XML file.")

    # Detect faces
    face_rects = face_cascade.detectMultiScale(face_img)

    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 5)

    return face_img
```

```python
result = detect_face(withglass)
```

```python
plt.imshow(result,cmap='gray')
plt.show()
```

```python
result = detect_face(group)
plt.imshow(result,cmap='gray')
plt.show()
```

```python
import cv2

def adj_detect_face(img):
    face_img = img.copy()

    # Load Haar cascade with the correct path
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    if face_cascade.empty():
        raise IOError("Haar cascade file not found or failed to load.")

    # Detect faces
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 10)

    return face_img
```

```python
# Doesn't detect the side face.
result = adj_detect_face(group)
plt.imshow(result,cmap='gray')
plt.show()
```

## Eye Cascade File

```python
eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')
```

```python
import cv2

def detect_eyes(img):
    face_img = img.copy()

    # Load Haar cascade for eyes
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
    
    if eye_cascade.empty():
        raise IOError("Cannot load Haar cascade for eye detection.")

    # Detect eyes
    eyes = eye_cascade.detectMultiScale(face_img)

    for (x, y, w, h) in eyes:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 1)

    return face_img
```

```python
result = detect_eyes(model)
plt.imshow(result,cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the eye cascade
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

# Check if the cascade loaded successfully
if eye_cascade.empty():
    raise IOError("Failed to load eye cascade")

# Load the image
img = cv2.imread('image_02.png')  # Replace with your actual image path

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect eyes
eyes = eye_cascade.detectMultiScale(gray)

# Print eye coordinates
print("Detected eye coordinates:")
for (x, y, w, h) in eyes:
    print(f"X: {x}, Y: {y}, Width: {w}, Height: {h}")
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(img_rgb)
plt.title("Detected Eyes")
plt.axis("off")
plt.show()
```

## Conjunction with Video

```python
import cv2
import matplotlib.pyplot as plt

# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Function to detect faces and draw rectangles
def detect_face(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
    
    return frame

# Open webcam
cap = cv2.VideoCapture(0)

# Check if webcam is opened
if not cap.isOpened():
    print("Cannot open camera")
    exit()

# Set up interactive plot
plt.ion()
fig, ax = plt.subplots()

# Read first frame
ret, frame = cap.read()
if not ret:
    print("Can't receive frame (stream end?). Exiting ...")
    cap.release()
    exit()

frame = detect_face(frame)
im = ax.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title("Video Face Detection")
plt.axis('off')

# Loop for video display
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = detect_face(frame)

    # Update the Matplotlib image
    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.draw()
    plt.pause(0.01)  # Lower this for faster refresh

# Release resources
cap.release()
plt.ioff()
plt.close()
```
